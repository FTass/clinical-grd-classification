```python
import pandas as pd
```

```python
df = pd.read_csv('../data/raw/dataset_elpino.csv', sep=';')
df.head()
```

```python
import sys
sys.path.append("..")


from src.data.EDA import EDA
eda = EDA()
eda.summaryEda(df, 'GRD')
```

```python
df_categoria = df.copy()


df_categoria["Diag 01 Principal (cod+des)"] = (
    df_categoria["Diag 01 Principal (cod+des)"]
    .str.split(" - ").str[0]   # A41.8
    .str[:3]                   # A41
)
df_categoria

col_diag = [c for c in df.columns if 'Diag' in c and "(cod+des)"in c]
col_proc= [c for c in df.columns if 'Proced' in c and "(cod+des)"in c]

for col in col_proc:
    df_categoria[col] = (
                        df_categoria[col].fillna('').astype(str)
                        .str.split(" - ").str[0].str[:2]
                        )
for col in col_diag:
    df_categoria[col] = (
        df_categoria[col].str.split(" - ").str[0].str[:3]
    )
df_categoria
df_categoria['Sexo (Desc)'].unique()
df_categoria['Sexo (Desc)'] = df_categoria['Sexo (Desc)'].map({'Hombre': 1, 'Mujer': 0})
df_categoria

df_categoria.to_csv('../data/processed/elpino_categories.csv', index=False)




```
