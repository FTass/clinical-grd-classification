Durante la fase de preparación y análisis estructural de los datos, se tomaron las siguientes decisiones metodológicas con el objetivo de asegurar la calidad, consistencia y trazabilidad del dataset:

Se creó una copia de trabajo del dataset original, con el fin de preservar la integridad de los datos fuente y permitir iteraciones seguras durante el proceso de limpieza y transformación.
Se identificó que el valor "-" en las columnas clínicas (diagnósticos y procedimientos) representaba ausencia de información, por lo que fue tratado como valor faltante (NaN) para su correcto manejo en el análisis.
Se realizó la separación de columnas compuestas que contenían tanto código como descripción (formato cod+des) en dos variables independientes, permitiendo un tratamiento más preciso de la información clínica.
Se verificó que el proceso de transformación no generó pérdida de registros, asegurando la consistencia del número de filas entre el dataset original y el dataset transformado.
Se evaluó la completitud de los datos tanto a nivel de posición (diagnósticos y procedimientos) como a nivel de bloque (diagnósticos, procedimientos, variables demográficas y variable objetivo), permitiendo caracterizar la densidad real de la información disponible.

A partir del análisis exploratorio y estructural del dataset, se identificaron las siguientes limitaciones y desafíos que deberán ser considerados en las etapas posteriores de modelado:

Se observa una alta sparsity en las posiciones avanzadas de diagnósticos secundarios, donde la completitud disminuye significativamente a medida que aumenta la posición, generando una gran cantidad de valores faltantes en columnas profundas.
Aunque los procedimientos presentan una mayor densidad relativa en comparación con los diagnósticos, también exhiben una disminución progresiva de completitud en posiciones avanzadas, lo que introduce heterogeneidad en la disponibilidad de información clínica.
El dataset original contenía múltiples columnas compuestas en formato código + descripción, lo que implicó un proceso adicional de transformación y puede introducir complejidad en la representación de variables.
Existe una gran cantidad de columnas altamente vacías (sobre 95% y 99% de valores nulos), lo que puede afectar negativamente el rendimiento de ciertos modelos si no se maneja adecuadamente la selección o reducción de variables.
La estructura del dataset, basada en posiciones (diagnósticos y procedimientos secuenciales), puede no ser la representación óptima para modelos de aprendizaje automático, por lo que podría requerirse una transformación hacia representaciones más compactas o informativas.
Se anticipa la necesidad de aplicar técnicas de reducción de dimensionalidad o selección de variables, con el fin de mitigar el efecto de la alta dispersión y mejorar la eficiencia del modelado.
Finalmente, aunque las variables demográficas y la variable objetivo presentan completitud total, su capacidad explicativa por sí sola es limitada, por lo que el desempeño del modelo dependerá fuertemente de cómo se representen los bloques clínicos.

En esta fase se realizó la preparación estructural del dataset, donde se transformaron los valores faltantes representados como "-" en valores nulos reales y se separaron las variables clínicas originalmente compuestas en código y descripción, manteniendo la lógica posicional del registro. Asimismo, se trabajó sobre una copia del dataset original, asegurando la integridad de los datos sin pérdida de filas durante el proceso de transformación.

A partir del análisis exploratorio, se confirmó que el dataset presenta una estructura jerárquica en los bloques clínicos, donde los diagnósticos muestran una disminución progresiva en su completitud a medida que aumenta la posición, mientras que los procedimientos presentan una mayor densidad relativa en múltiples niveles. También se validó que las variables demográficas y la variable objetivo se encuentran completamente disponibles.

Sin embargo, persisten desafíos relevantes asociados a la alta sparsity en posiciones avanzadas, la existencia de múltiples columnas altamente vacías y la necesidad de definir una representación adecuada para los bloques clínicos, dado que la estructura posicional no necesariamente es óptima para el modelado.

```python
import pandas as pd
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder

# =========================
# 1. Base final para modelar
# =========================
X = df_model.copy()

if "sexo" in X.columns:
    X = X.drop(columns=["sexo"])

y = df_diag_split["GRD"].copy()

# =========================
# 2. Definir columnas
# =========================
cols_numericas = [
    "n_diag",
    "n_proc",
    "has_proc",
    "multi_diag",
    "edad",
    "sexo_bin"
]

cols_categoricas = [
    "diag_principal",
    "proc_principal",
    "edad_grupo"
]

# =========================
# 3. Preprocesador
# =========================
preprocessor = ColumnTransformer(
    transformers=[
        (
            "cat",
            OneHotEncoder(
                handle_unknown="infrequent_if_exist",
                min_frequency=20,
                sparse_output=False
            ),
            cols_categoricas
        ),
        (
            "num",
            "passthrough",
            cols_numericas
        )
    ]
)

# =========================
# 4. Fit + transform
# =========================
X_encoded = preprocessor.fit_transform(X)

# =========================
# 5. Nombres de columnas
# =========================
cat_feature_names = preprocessor.named_transformers_["cat"].get_feature_names_out(cols_categoricas)
feature_names = list(cat_feature_names) + cols_numericas

# =========================
# 6. DataFrame final
# =========================
X_encoded_df = pd.DataFrame(
    X_encoded,
    columns=feature_names,
    index=X.index
)

X_final = X_encoded_df
y_final = y

# =========================
# 7. Validaciones
# =========================
print("Shape original X:", X.shape)
print("Shape encoded X:", X_final.shape)
print("Nulos en X_final:", X_final.isna().sum().sum())

print("\nPrimeras filas:")
print(X_final.head())
```

```python
y_grouped = y_final.astype(str).str[:3]
```

```python
print("Clases originales:", y_final.nunique())
print("Clases agrupadas:", y_grouped.nunique())
```

```python
conteo_grouped = y_grouped.value_counts()

print("Clases agrupadas totales:", y_grouped.nunique())
print("Clases con 1 caso:", (conteo_grouped == 1).sum())
print("Clases con menos de 2 casos:", (conteo_grouped < 2).sum())
print("Clases con menos de 5 casos:", (conteo_grouped < 5).sum())
```

```python
clases_validas = conteo_grouped[conteo_grouped >= 2].index

mask = y_grouped.isin(clases_validas)

X_grouped = X_final[mask].copy()
y_grouped_filtrado = y_grouped[mask].copy()

print("Shape original:", X_final.shape, y_grouped.shape)
print("Shape filtrado:", X_grouped.shape, y_grouped_filtrado.shape)
print("Nuevas clases:", y_grouped_filtrado.nunique())
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X_grouped,
    y_grouped_filtrado,
    test_size=0.2,
    random_state=42,
    stratify=y_grouped_filtrado
)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score

model = RandomForestClassifier(
    n_estimators=100,
    random_state=42,
    n_jobs=-1
)

model.fit(X_train, y_train)

y_pred = model.predict(X_test)

print("Accuracy agrupado filtrado:", accuracy_score(y_test, y_pred))
```

```python
from sklearn.metrics import classification_report

print(classification_report(y_test, y_pred))
```

```python
from sklearn.metrics import precision_score, recall_score, f1_score

print("Precision (macro):", precision_score(y_test, y_pred, average="macro"))
print("Recall (macro):", recall_score(y_test, y_pred, average="macro"))
print("F1 (macro):", f1_score(y_test, y_pred, average="macro"))

print("\nPrecision (weighted):", precision_score(y_test, y_pred, average="weighted"))
print("Recall (weighted):", recall_score(y_test, y_pred, average="weighted"))
print("F1 (weighted):", f1_score(y_test, y_pred, average="weighted"))
```
